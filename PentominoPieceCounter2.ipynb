```python
# with open('data.txt', 'r') as file:
#     data = file.read().rstrip()
with open("/content/SampleSolution1.txt","r") as lines:
  lines=lines.read().rstrip()
  #print(lines)
  lines = lines.replace("4","Y").replace("10","N").replace("11","W").replace("12","X").replace("8","U").replace("7","P").replace("9","T").replace("1","Z").replace("2","V").replace("3","F").replace("5","L").replace("6","I").replace("\n", "").replace("\r", "").replace(",", "")
  #print(lines)
m=1
n=61
tiling_dict={}
string_temp=("")
# list_temp=[]
count = 0

while count < 8:
  string_temp = lines[m:n]
  # count2 = 0
  # for item in list_temp:
  #   item_str = str(item) + str(count2)
  #   list_temp.append(item_str)
  #   count2 = count2 + 1
  tiling_dict[count]=string_temp
  m=n+1
  n=n+61
  count=count+1
print(tiling_dict)

```

```python
count_dict=0
while count_dict < 8:
  tiling_list=[]
  count = 0
  for item in tiling_dict[count_dict]:
    # print(item)
    item_str = str(item) + str(count)
    tiling_list.append(item_str)
    count = count + 1
  # print(tiling_list)
  tiling_list.sort()
  tiling_dict[count_dict] = tiling_list
  #print(tiling_list)
  # print(tiling_dict[count_dict])
  count_dict = count_dict + 1
print(tiling_dict)


```

```python
piece_position_dict={}
count = 0
while count<8:
  F=tiling_dict[count][0:5]
  I=tiling_dict[count][5:10]
  L=tiling_dict[count][10:15]
  N=tiling_dict[count][15:20]
  P=tiling_dict[count][20:25]
  T=tiling_dict[count][25:30]
  U=tiling_dict[count][30:35]
  V=tiling_dict[count][35:40]
  W=tiling_dict[count][40:45]
  X=tiling_dict[count][45:50]
  Y=tiling_dict[count][50:55]
  Z=tiling_dict[count][55:60]

  Fstr="'".join(F).replace("'", "")
  Istr="'".join(I).replace("'", "")
  Lstr="'".join(L).replace("'", "")
  Nstr="'".join(N).replace("'", "")
  Pstr="'".join(P).replace("'", "")
  Tstr="'".join(T).replace("'", "")
  Ustr="'".join(U).replace("'", "")
  Vstr="'".join(V).replace("'", "")
  Wstr="'".join(W).replace("'", "")
  Xstr="'".join(X).replace("'", "")
  Ystr="'".join(Y).replace("'", "")
  Zstr="'".join(Z).replace("'", "")
  pieceposition_tiling_list = [Fstr,Istr,Lstr,Nstr,Pstr,Tstr,Ustr,Vstr,Wstr,Xstr,Ystr,Zstr]
  piece_position_dict[count]=pieceposition_tiling_list
  print(pieceposition_tiling_list)
  count = count + 1

```

```python
sort_pieces_list=[]
count = 0
while count<8:
  sort_pieces_list = piece_position_dict[count] + sort_pieces_list
  # print(sort_pieces_list)
  sort_pieces_list.sort()
  # print(sort_pieces_list)
  count=count+1
print(sort_pieces_list)

```
